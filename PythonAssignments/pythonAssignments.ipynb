<font size="+1">**This program accepts a person's Blood Sugar Level and provides a category based on the result.**<font>

<font size="+1">**Use the Table as a reference for calculations**<font>

**Input**
  * Fasting Sugar Level (in mg/dl)
  * Post Meal Sugar Level (in mg/dl)

***Use the diagram below as a reference***

image.png

```python
#
# Accept values from the user
#
sugarLevelTiming  = input("Is this a 'Fasting' or 'Post-Meal' Reading : ").upper()
sugarLevelReading = float(input("Enter your Blood Sugar Level (in mg/dL)    : "))

print("===========================================================")
print(f"Your Sugar Level Timing   : {sugarLevelTiming}"      )
print(f"Your Sugar Level Reading  : {sugarLevelReading:.2f}" )
print("===========================================================")

if sugarLevelTiming != 'FASTING' and sugarLevelTiming != 'POST-MEAL':
  print("ERROR : INVALID TIMING. Please enter either 'Fasting' or 'Post-Meal'")
else:
  #
  # Calculate the Category based on the Timing and Sugar Level
  #
  if sugarLevelTiming == "FASTING":
    if sugarLevelReading < 70:
      sugarLevelCategory = "Low Blood Sugar (Hypoglycemia)"
    elif 70 <= sugarLevelReading <= 100:
      sugarLevelCategory = "Normal (Fasting)"
    elif 101 <= sugarLevelReading <= 125:
      sugarLevelCategory = "Pre-Diabetic (Fasting)"
    elif sugarLevelReading > 125:
      sugarLevelCategory = "Diabetic (Fasting)"
  elif sugarLevelTiming == 'POST-MEAL':
    if sugarLevelReading < 70:
      sugarLevelCategory = "Low Blood Sugar (Hypoglycemia)"
    elif 70 <= sugarLevelReading <= 140:
      sugarLevelCategory = "Normal (Post-Meal)"
    elif 141 <= sugarLevelReading <= 200:
      sugarLevelCategory = "Pre-Diabetic (Post-Meal)"
    elif sugarLevelReading > 200:
      sugarLevelCategory = "Diabetic (Post-Meal)"

  print(f"Your Sugar Level Category : {sugarLevelCategory}"          )
  print("===========================================================")
```

<font size="+2">**Method 1 - Dictionary of Dictionaries**<font>


---


<font size="+1">**This program receives a user-specified count and then populates a dictionary with the corresponding number of country key-value pairs**<font>

<font size="+1">Creates a single dictionary where each country's name is the **key**. The **value** for each key is another dictionary containing the country's population, area, and capital<font>


**Input**
* Country Name
* Population
* Area
* Capital

```python
#
# Get the number of Countries from the User.
#

#
# Initialize a dictionary to store all the country data, with the country name as the key.
#
dictCountry   = dict()
dictCountries = dict()

try:
  numCountries = int(input("Enter the Number of Countries you want to create the details for : "))
except ValueError:
  print("\nInvalid input. Please enter a number greater than 0.")
  exit()

for i in range(numCountries):
  #
  # Get Country Name
  #
  invalidValue = True

  while invalidValue:
    invalidValue = False
    countryName = input("Enter the Country Name : ")

    #
    # Check if the string is empty or just whitespace
    #
    if not countryName.strip():
      invalidValue = True
      print("\nInvalid Country Name. Country Name Cannot be Empty.")

  #
  # Get Country Population
  #
  invalidValue = True

  while invalidValue:
    try:
      invalidValue = False
      countryPopulation = int(input("Enter the Population : "))

      #
      # Check if the Population is valid
      #
      if countryPopulation < 1:
        invalidValue = True
        print("\nInvalid Population. Please enter a number greater than 0.")
        continue # Restart the loop
    except ValueError:
      invalidValue = True
      print("\nInvalid Population. Please enter a number greater than 0.")
      continue # Restart the loop

  #
  # Get Country Area
  #
  invalidValue = True

  while invalidValue:
    try:
      invalidValue = False
      countryArea = int(input("Enter the Area : "))

      #
      # Check if the Area is valid
      #
      if countryArea < 1:
        invalidValue = True
        print("\nInvalid Area. Please enter a number greater than 0.")
        continue # Restart the loop
    except ValueError:
      invalidValue = True
      print("\nInvalid Area. Please enter a number greater than 0.")
      continue # Restart the loop
  #
  # Get Capital
  #
  invalidValue = True

  while invalidValue:
    invalidValue = False
    countryCapital = input("Enter the Capital : ")

    #
    # Check if the string is empty or just whitespace
    #
    if not countryCapital.strip():
      invalidValue = True
      print("\nInvalid Capital. Capital Name Cannot be Empty.")

  print("\n" + "~" * 75)

  dictCountry = {
                  'name'       : countryName,
                  'population' : countryPopulation,
                  'area'       : countryArea,
                  'capital'    : countryCapital
                }

  #
  # Add the dictionary to the main dictionary, with the country name as the key.
  #
  dictCountries[countryName] = dictCountry

print(dictCountries)

#
# Print the final dictionary of all countries for confirmation
#
print("\n" + "=" * 75)
print("All Countries Entered")
print("=" * 75)
for countryName, countryDetails in dictCountries.items():
    print(f"Country Name : {countryName}")
    print(f"  Population : {countryDetails['population']}")
    print(f"  Area       : {countryDetails['area']}")
    print(f"  Capital    : {countryDetails['capital']}")

print("\nProgram finished successfully.")
```

<font size="+2">**Method 2 - Dictionary with Lists**<font>


---


<font size="+1">**This program receives a user-specified count and then populates a dictionary with the corresponding number of country key-value pairs**<font>

<font size="+1">This program uses a single dictionary where each **key** represents a data category (`name`, `population`, `area`, `capital`) and the **value** is a list of all the entries for that category<font>

**Input**
* Country Name
* Population
* Area
* Capital

```python
#
# Initialize a dictionary with lists to store all the country data.
#
dictCountries = {
                  'name': [],
                  'population': [],
                  'area': [],
                  'capital': []
                }

try:
  numCountries = int(input("Enter the Number of Countries you want to create the details for : "))
except ValueError:
  print("\nInvalid input. Please enter a number greater than 0.")
  exit()

for i in range(numCountries):
  #
  # Get Country Name
  #
  invalidValue = True

  while invalidValue:
    invalidValue = False
    countryName = input("Enter the Country Name : ")

    #
    # Check if the string is empty or just whitespace
    #
    if not countryName.strip():
      invalidValue = True
      print("\nInvalid Country Name. Country Name Cannot be Empty.")

  #
  # Get Country Population
  #
  invalidValue = True

  while invalidValue:
    try:
      invalidValue = False
      countryPopulation = int(input("Enter the Population : "))

      #
      # Check if the Population is valid
      #
      if countryPopulation < 1:
        invalidValue = True
        print("\nInvalid Population. Please enter a number greater than 0.")
        continue # Restart the loop
    except ValueError:
      invalidValue = True
      print("\nInvalid Population. Please enter a number greater than 0.")
      continue # Restart the loop

  #
  # Get Country Area
  #
  invalidValue = True

  while invalidValue:
    try:
      invalidValue = False
      countryArea = int(input("Enter the Area : "))

      #
      # Check if the Area is valid
      #
      if countryArea < 1:
        invalidValue = True
        print("\nInvalid Area. Please enter a number greater than 0.")
        continue # Restart the loop
    except ValueError:
      invalidValue = True
      print("\nInvalid Area. Please enter a number greater than 0.")
      continue # Restart the loop
  #
  # Get Capital
  #
  invalidValue = True

  while invalidValue:
    invalidValue = False
    countryCapital = input("Enter the Capital : ")

    #
    # Check if the string is empty or just whitespace
    #
    if not countryCapital.strip():
      invalidValue = True
      print("\nInvalid Capital. Capital Name Cannot be Empty.")

  print("\n" + "~" * 75)

  #
  # Append the collected data to the respective lists in the main dictionary.
  #
  dictCountries['name'].append(countryName)
  dictCountries['population'].append(countryPopulation)
  dictCountries['area'].append(countryArea)
  dictCountries['capital'].append(countryCapital)

#
# Print the final dictionary of all countries for confirmation
#
print("\n" + "=" * 75)
print("All Countries Entered")
print("=" * 75)

print(dictCountries)
print(dictCountries.items())

for k, v in dictCountries.items():
  print(k, v)

print("\n" + "=" * 75)
print("All Countries Entered")
print("=" * 75)

# for countryName, countryPopulation, countryArea, countryCapital in dictCountries.items():
#
# The above raises this Error
#
# ValueError                                Traceback (most recent call last)
# /tmp/ipython-input-3767996266.py in <cell line: 0>()
#     111 print("=" * 75)
#     112
# --> 113 for countryName, countryPopulation, countryArea, countryCapital in dictCountries.items():
#     114     print(f"Country Name : {countryName}")
#     115     print(f"  Population : {countryPopulation}")
#
# ValueError: not enough values to unpack (expected 4, got 2)
#
# The Cause of the Error
# The dict.items() method returns a list of key-value pairs. For your countries_dict, the items are:
# ('name', ['India', 'United Kingdom', ...])
# ('population', [1460000000, 69230000, ...])
# ('area', [3287263, 243610, ...])
# ('capital', ['New Delhi', 'London', ...])
#
# The for loop is trying to unpack each pair into four variables, but it's only receiving a key and a value (two items) at a time
# This mismatch between the four variables you expect and the two values you receive is what causes the ValueError
#
# To fix this, we need a way to iterate over all the lists within the dictionary at the same time
# The best way to do this in Python is with the built-in zip() function
# zip() takes multiple iterable objects and combines them element by element

for countryName, countryPopulation, countryArea, countryCapital in zip(dictCountries['name'], dictCountries['population'], dictCountries['area'], dictCountries['capital']):
    print(f"Country Name : {countryName}")
    print(f"  Population : {countryPopulation}")
    print(f"  Area       : {countryArea}")
    print(f"  Capital    : {countryCapital}")

print("\nProgram finished successfully.")
```
